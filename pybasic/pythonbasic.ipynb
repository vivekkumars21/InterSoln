### Question 1: Largest of Three Numbers

```python
num1 = int(input('Enter Number 1'))
num2 = int(input('Enter number 2'))
num3 = int(input('Enter number 3'))
if num1 >= num2 and num1 >= num3:
    print(num1)
elif num2 >= num1 and num2 >= num3:
    print(num2)
else:
    print(num3)
```

### Question 4: Function to Count Vowels

```python
def count_vowels(s):
    vowels = "aeiouAEIOU"
    count = sum(1 for char in s if char in vowels)
    return count

s = input('Enter name')
print(count_vowels(s))
```

### Question 5: Match Case - Month Days

```python
month = int(input("enter months"))
match month:
    case 2:
        print(28)
    case 4 | 6 | 9 | 11:
        print(30)
    case _:
        print(31)
```

### Question 7: Function with Multiple Parameters

```python
def calculate_salary(basic, hra, bonus):
    return basic + hra + bonus

basic = int(input("base salary"))
hra = int(input("enter Hra"))
bonus = int(input('Enter Bonus'))
print(calculate_salary(basic, hra, bonus))
```

### Question 8: Employee Bonus Calculator

```python
def calculate_bonus(salary, experience):
    if experience >= 10:
        bonus = 0.20 * salary
    elif experience >= 5:
        bonus = 0.10 * salary
    else:
        bonus = 0.05 * salary
    return bonus

salary = int(input('Salary  = '))
experience = int(input('exp = '))
print(calculate_bonus(salary, experience))
```

### Question 9: Student Result Analyzer

```python
def calculate_result(mark1, mark2, mark3):
    for mark in [mark1, mark2, mark3]:
        if mark < 0 or mark > 100:
            return "Invalid Marks"
    avg = (mark1 + mark2 + mark3) / 3
    if avg >= 75:
        return "Distinction"
    elif avg >= 60:
        return "First Class"
    elif avg >= 40:
        return "Second Class"
    else:
        return "Fail"

try:
    mark1 = int(input('Enter marks 1'))
    mark2 = int(input('Enter marks 2'))
    mark3 = int(input('Enter marks 3'))
    print(calculate_result(mark1, mark2, mark3))
except ValueError:
    print("Invalid Marks")
```
